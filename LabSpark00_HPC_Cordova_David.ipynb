## Apache Spark
#### • Spark es un FrameWork para Big Data que incluye bibliotecas para el análisis de datos, el aprendizaje automático, el análisis de grafos y el procesamiento de datos en tiempo real.
#### • Spark se presenta mucho más rápido dado que escribe los resultados/objetos intermedios en memoria principal siempre que sea posible.
#### • El ecosistema de Hadoop incluye un sistema de almacenamiento de archivos distribuido llamado HDFS (Hadoop Distributed File System).
#### • Spark, por su parte, no incluye un sistema de almacenamiento de archivos.
#### • Se puede utilizar Spark sobre HDFS, o de otras fuentes (locales o en la nube).
#### • Spark no es un lenguaje de programación como Python o Java. Se trata de un motor de procesamiento de datos distribuido de uso general, adecuado para su aplicación en una amplia variedad de situaciones. Resulta especialmente útil para el procesamiento de big data, tanto a gran escala como a alta velocidad.
#### • Los desarrolladores de aplicaciones y los científicos de datos suelen incorporar Spark en sus aplicaciones para consultar, analizar y transformar datos rápidamente y a gran escala.
#### • Algunas de las tareas que se asocian con mayor frecuencia a Spark incluyen: –Trabajos por lotes de ETL y SQL en grandes conjuntos de datos (a menudo de terabytes de tamaño). – Procesamiento de datos en streaming procedentes de dispositivos y nodos de IoT, datos de diversos sensores, sistemas financieros y transaccionales de todo tipo. – Tareas de aprendizaje automático para aplicaciones de comercio electrónico o de TI.
####  • En esencia, Spark se basa en el marco Hadoop/HDFS para gestionar archivos distribuidos. Se implementa principalmente con Scala, una variante del lenguaje funcional de Java.
#### • Existen bibliotecas desarrolladas para el análisis de consultas de tipo SQL, el aprendizaje automático distribuido, el cálculo de grafos a gran escala y el procesamiento de datos en streaming, que se pueden acceder desde SPARK.
#### • Spark admite múltiples lenguajes de programación en forma de bibliotecas de interfaz fáciles de usar: Java, Python, Scala y R.

```python
### Estas celdas se encargan de crear los conjuntos de datos iniciales, generalmente en forma de listas o RDDs.
###--> Creación de sesión SPARK
configura = SparkConf()
###--> Se crea la configuración
configura.setAppName("Spark_Cordova_00")
###--> Se crea la SESIÓN en función de la configuración (Spark+Apellido del autor)
sparkCordova = SparkSession.builder.config(conf=configura).getOrCreate()
###--> Se crea el contexto de SPARK para consultas SQL
SQLContext(sparkContext=sparkCordova.sparkContext, sparkSession=sparkCordova)
###-->Se crea el contexto de SPARK GENERAL
sparkContextoCordova= sparkCordova.sparkContext.getOrCreate()
###-->Se imprime la sesion
print("Sesion creada: PySpark_HPC_00")
###-->
sparkCordova
```

```python
### Comentario:
## En esta celda se realiza la creación de RDDs vacíos utilizando dos métodos diferentes.
## En el primer caso, se emplea la función emptyRDD() para generar un RDD sin elementos.
## En el segundo caso, se utiliza parallelize() sobre una lista vacía para obtener el mismo resultado.
## Posteriormente, se aplica la función isEmpty() en ambos casos, la cual retorna un valor verdadero, confirmando que los RDDs no contienen datos. Esto permite validar que ambos enfoques son equivalentes para la creación de estructuras vacías en Spark.
###--> 3.1.- Creación de un RDD vacio
rdd = sparkContextoCordova.emptyRDD()
rdd.isEmpty()
###-->
a = sparkContextoCordova.parallelize([])
a.isEmpty()
```

```python
### Comentario:
### Estas celdas se crea un RDD con distintos tipos de datos mediante parallelize(), lo que demuestra que puede almacenar estructuras heterogéneas. Luego, con collect() se visualiza su contenido, confirmando que los datos se cargaron correctamente.
### Finalmente, getNumPartitions() permite verificar el número de particiones del RDD, mostrando cómo Spark distribuye los datos internamente.
###--> 3.3.- Se crea un RDD con particiones manualmente
mixRDD = sparkContextoCordova.parallelize([True, [11,22,33,44,55], (10,20,30,40,50)])
mixRDD.collect()
###--> 3.4.- Se revisa el número de particiones
mixRDD.getNumPartitions()
```

```python
### Se crea un RDD con datos de distintos tipos usando parallelize().
### collect() permite visualizar los datos cargados correctamente.
### getNumPartitions() muestra cómo se distribuyen los datos en particiones.
###--> 3.5.- Se recopila los datos con particiones
mixRDD.glom().collect()
###--> Se le pone un nombre
mixRDD.setName('Mi_RDD')
###--> Usando Rangos para crear RDD
a= sparkContextoCordova.parallelize(range(1,4000), 4)
print(a.getNumPartitions())
print("\n")
print(a.glom().take(1)) ###-->
print("\n")
print(a.glom().max()) #https://www.databricks.com/blog/2015/06/22/understanding-yoprint("\n")
print(a.glom().min()) #
print("\n")
```

```python
### Se crea un RDD con valores numéricos.
### takeOrdered() ordena los datos usando una función lambda.
### La lambda invierte el orden para obtener los valores más altos.
### Se retornan los 4 valores mayores del RDD.
###--> Se ordenan y re-reparten RDD
###--> Uso de función lambda o función anónima para ordenar
b = sparkContextoCordova.parallelize([12,21,23,43,1,22,11,45,56])
b.takeOrdered(4, key=lambda x: -x)
```

```python
### Se crea un RDD con valores del 1 al 999 dividido en 5 particiones.
### getNumPartitions() confirma que el RDD tiene 5 particiones.
### Permite analizar la distribución de datos dentro del procesamiento en Spark.
###--> Coalescence, reparticiones y debug (DAG)
c = sparkContextoCordova.parallelize(range(1,1000), 5)
###--> Obteniendo el numero de particiones
c.getNumPartitions()
```

```python
### Se usa repartition() para cambiar el número de particiones del RDD.
### Primero se aumenta a 7 particiones y luego se reduce a 4.
### getNumPartitions() verifica que los cambios se aplicaron correctamente.
###--> Poniendole 7 particiones
d = c.repartition(7)
d.getNumPartitions()
##--> Poniendole 4 particiones
w = c.repartition(4)
w.getNumPartitions()
```

```python
!hdfs dfs -ls /
```

```python
# Se usa coalesce() para reducir el número de particiones del RDD.
# Primero se ajusta a 6 particiones y luego a 2.
# getNumPartitions() confirma los cambios realizados.
# saveAsTextFile() guarda el RDD en almacenamiento para su uso o análisis posterior.

# --> Usando coalesce para 6 particiones
# coalesce(6) reduce las particiones actuales a 6 sin hacer shuffle de datos
w = w.coalesce(6)
# Verifica que el RDD ahora tiene 6 particiones
w.getNumPartitions()

# --> Usando un coalesce de 2
# Reduce nuevamente las particiones a 2 para mayor consolidación
w = w.coalesce(2)
# Verifica que el RDD ahora tiene 2 particiones
w.getNumPartitions()

# --> Almacenar RDD y debugging
# os y shutil permiten manipular el sistema de archivos desde Python
import os, shutil

# Define la ruta de salida donde se guardará el RDD
output_path = "/tmp/Guardar01"

# Spark no sobreescribe directorios existentes, por eso se elimina
# el directorio previo si ya existe antes de guardar
if os.path.exists(output_path):
    shutil.rmtree(output_path)  # Elimina el directorio y todo su contenido

# coalesce(1) consolida todo el RDD en una sola partición
# para que se guarde en un único archivo de texto
w.coalesce(1).saveAsTextFile("file:///tmp/Guardar01")
```

## 4.- Operaciones sobre los RDD's

## 5.- Creación de Funciones Definidas por el Usuario (UDF)

```python
## --> Creación y visualización de un RDD de palabras
## Se define una lista de strings, se convierte en un RDD
## y se recuperan sus elementos utilizando collect()

palabras = ['Que', 'tal', 'estas', 'hola', 'hey', 'hora']

palabraRDD = sparkContextoCordova.parallelize(palabras)

# Mostrar todos los elementos del RDD
palabraRDD.collect()
```

```python
## --> Función de filtrado por inicial
## Verifica si una palabra comienza con la letra 'h' (sin importar mayúsculas/minúsculas)
def inicio_h(palabras):
    return palabras[0].lower().startswith('h')
```

```python
## --> Filtrado de palabras que comienzan con 'h'
## Se aplica la función definida previamente para filtrar el RDD
## y se muestran los resultados con collect()

palabraRDD.filter(inicio_h).collect()
```

```python
## --> Transformación a mayúsculas
## Convierte cada palabra del RDD a letras mayúsculas

def mayuscula(p):
    return p.upper()
```

```python
## --> Transformación de un RDD a mayúsculas
## Se crea un RDD a partir de una lista de palabras y se aplica una función
## de transformación (map) para convertir cada elemento a mayúsculas.
## Finalmente, se muestran los resultados con collect()

lista0 = ['cómputo', 'mayor', 'menor']

lista1 = sparkContextoCordova.parallelize(lista0)

# Aplicar transformación a mayúsculas
mapeo0 = lista1.map(mayuscula)

# Mostrar resultados
mapeo0.collect()
```

```python
## --> Transformación a mayúsculas usando función lambda
## Se aplica map() con una función anónima (lambda) para convertir
## cada palabra del RDD a mayúsculas y se muestran los resultados

mapeo1 = lista1.map(lambda k: k.upper())

mapeo1.collect()
```

```python
## --> Carga y visualización de datos desde HDFS
## Se lee el archivo test.csv ubicado en HDFS como un RDD de texto,
## donde cada línea del fichero corresponde a un elemento del RDD.
## Posteriormente, se recupera el contenido completo con collect().

fichero = sparkContextoCordova.textFile("hdfs://10.195.34.34:9000/csv/titanic/test.csv")

# Mostrar el contenido del archivo
fichero.collect()
```

```python
## --> Procesamiento de datos tipo CSV
## - flatMap: divide cada línea en elementos individuales usando ","
## - map: convierte cada elemento en una tupla (valor, 1)
## - reduceByKey: suma los valores para contar ocurrencias
## - collect: devuelve el resultado al driver

fichero.flatMap(lambda x: x.split(",")) \
       .map(lambda x: (x, 1)) \
       .reduceByKey(lambda a, b: a + b) \
       .collect()
```

## 7.- Aplicaciones de funciones varias: map, parallelize y lambda

```python
## --> Ejemplo de transformación de estructura de datos
## Cada número del RDD se convierte en una lista con tres copias del mismo valor.
## Esto demuestra cómo map() puede cambiar la forma de los datos.

sparkContextoCordova.parallelize([4,5,6,7]) \
    .map(lambda x: [x, x, x]) \
    .collect()
```

```python
## Ejemplo de transformación map que genera estructuras anidadas:
## A partir de un RDD de números, cada elemento se transforma en una lista
## que contiene: una lista con repeticiones del número, otra con su cubo
## y otra con una operación adicional (suma). Ilustra cómo map puede producir
## listas complejas dentro de cada elemento del RDD.
###--->
sparkContextoCordova.parallelize([5,6,7,8]).map(lambda x:[[x,x,x],[x*x*x],[x+5]])
```

```python
## Ejemplo de acciones sobre RDD:
## Se crea un RDD con nombres de frutas y se utilizan acciones como collect()
## para recuperar los datos y count() para obtener la cantidad de elementos.
## Luego, se genera otro RDD numérico y se aplica countByValue(), que devuelve
## un diccionario con la frecuencia de cada valor, útil para análisis de conteo.
###--->
entradaFruta = sparkContextoCordova.parallelize(["manzana", "banana", "piña", "limón"])
###--->
print(f"{entradaFruta.collect()} \n La cantidad de frutas = {entradaFruta.count()}")
###---> Creación de un diccionario según la cantidad de valores que se repiten
dicPorCantidad = sparkContextoCordova.parallelize([2,2,2,4,4,4,4,6,6,6,6,6,7]).countByValue()
dicPorCantidad
```

### RDDs y particionamiento

#### En la primera parte se trabajo con RDDs, que son la estructura base de Spark. Se verifico que es posible crear RDDs vacios, con tipos de datos mixtos y con rangos numericos extensos. Se comprobo que Spark distribuye los datos en particiones de forma automatica y que el usuario puede controlar esa distribucion mediante `repartition()` y `coalesce()`. La diferencia entre ambos quedo clara: `repartition()` realiza un shuffle completo de los datos mientras que `coalesce()` solo reduce particiones sin redistribuir, lo que lo hace mas eficiente cuando se quiere disminuir el numero de particiones.

### Operaciones y transformaciones sobre RDDs

#### Se aplicaron operaciones de reduccion como `reduce()` con funciones lambda para calcular sumas y productos, lo que demostro que Spark ejecuta las operaciones de forma distribuida sin necesidad de traer los datos al driver hasta el momento de la accion final. Tambien se trabajo con transformaciones como `map()`, `flatMap()`, `filter()` y funciones definidas por el usuario, confirmando que Spark separa las transformaciones (lazy) de las acciones (que disparan la ejecucion real).

### Interaccion con HDFS

#### La interaccion con HDFS funciono correctamente. Se cargo el dataset del Titanic desde el sistema de archivos distribuido y se proceso con operaciones clasicas de MapReduce (`split`, `map` con tupla, `reduceByKey`), lo cual ilustra como Spark puede replicar y superar el paradigma original de Hadoop MapReduce con una sintaxis mucho mas compacta.

Se creo un DataFrame manualmente y luego se cargo el dataset de calidad de vino tinto desde HDFS. Fue necesario realizar un casting explicito de todas las columnas porque Spark lee los CSV como strings por defecto, lo cual es un paso que no se debe omitir en ningun flujo de trabajo real. Una vez con los tipos correctos, se obtuvieron estadisticas descriptivas, se seleccionaron columnas, se ordenaron registros y se verifico la ausencia de valores nulos.

### Filtros y slicing

Tambien se practico el slicing de filas y columnas con `limit()` y seleccion por indices, y la conversion a Pandas con `toPandas()` para los casos donde se necesita trabajar localmente con un subconjunto de datos.

### Observaciones generales

#### Spark resulta considerablemente mas agil que MapReduce clasico para tareas de exploracion y transformacion de datos. La combinacion de RDDs para procesamiento de bajo nivel y DataFrames para analisis estructurado cubre la mayoria de necesidades en un entorno de procesamiento de alto volumen. El uso de HDFS como capa de almacenamiento distribuido complementa bien el motor de Spark, permitiendo que los datos esten disponibles para todos los nodos del cluster sin necesidad de replicacion manual.
